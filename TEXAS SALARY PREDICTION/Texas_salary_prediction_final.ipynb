#### OUTLIERS

```python
plt.figure(figsize=(20,30),facecolor='white')
plotnumber=1

for column in continous_data:
    if plotnumber<=8:
        ax = plt.subplot(4,2,plotnumber)
        sns.boxplot(data= data, x=column, orient='h')
        plt.xlabel(column, fontsize= 13)
        plt.ylabel('count', fontsize= 13)
    plotnumber+=1
plt.tight_layout()
```

- From the above boxplot we can see that there are more outliers for the columns monthly,agency,hrly_rate,hrs_per_wk except for state_number

```python
# since the plot is skewed, IQR method is used for handling outliers

Q1 = continous_data.quantile(0.25)
Q3 = continous_data.quantile(0.75)

IQR = Q3-Q1

min_limit = Q1-1.5*IQR
max_limit = Q3+1.5*IQR

outliers_count = ((continous_data > max_limit) | (continous_data < min_limit)).sum()

outliers_percentage = (outliers_count/len(continous_data))*100

print('\n Sum of Outliers: \n',outliers_count)
print('\n Percentage of Outliers:\n',outliers_percentage)
```

- We are going to handle outliers for agency, hrly_rate, hrs_per_wk except for monthly as we can see the percentage of outliers is more than 5%

```python
# handling outliers for hrly_rate column

from scipy import stats

Q1 = np.percentile(continous_data.hrly_rate,25)
Q3 = np.percentile(continous_data.hrly_rate,75)

IQR = stats.iqr(continous_data['hrly_rate'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(continous_data['hrly_rate'] > max_limit) | (continous_data['hrly_rate'] < min_limit),'hrly_rate'] = np.median(continous_data['hrly_rate'])
```

```python
# handling outliers for hrs_per_wk column

Q1 = np.percentile(continous_data.hrs_per_wk,25)
Q3 = np.percentile(continous_data.hrs_per_wk,75)

IQR = stats.iqr(continous_data['hrs_per_wk'])

max_limit = Q3+(1.5*IQR)
min_limit = Q1-(1.5*IQR)

data.loc[(continous_data['hrs_per_wk'] > max_limit) | (continous_data['hrs_per_wk'] < min_limit),'hrs_per_wk'] = np.median(continous_data['hrs_per_wk'])
```

### (B) What departments/roles have the biggest wage disparities between managers and employees?

```python
# Find the model with the lowest MSE
best_model_mse = min(model_results, key=lambda x: x["MSE"])

# Find the model with the highest R2 score
best_model_r2 = max(model_results, key=lambda x: x["R2 Score"])

# Determine the best model based on both metrics
if best_model_mse == best_model_r2:
    best_model = best_model_mse
    print("Best Model:", best_model["Model"])
else:
    print("No unanimous best model. Consider trade-offs between MSE and R2 Score.")
```
